# Quest茫o 1

Implemente uma fun莽茫o `load_data(seed)` que:

- carregue o dataset CIFAR-10 utilizando `tensorflow.keras.datasets.cifar10.load_data`;
- realize o flatten das imagens;
- normalize os dados;
- realize a separa莽茫o entre treino e valida莽茫o;
- utilize `train_test_split` com controle de aleatoriedade (`seed`);
- retorne:

```python
X_train, X_val, y_train, y_val
```

j谩 normalizados e preparados para treinamento.

Al茅m disso, responda:

1. Qual o formato original das imagens?
2. Quantas features cada imagem possui ap贸s o flatten?
3. Por que o flatten 茅 necess谩rio para uma MLP?
4. Qual a import芒ncia da normaliza莽茫o para o treinamento?

**Solu莽茫o**:

````python
from tensorflow.keras.datasets import cifar10
from sklearn.model_selection import train_test_split

def load_data(seed=42):
    # Carrega o dataset
    (X_train_full, y_train_full), (X_test, y_test) = cifar10.load_data()

    # Flatten: (50000, 32, 32, 3) -> (50000, 3072)
    X_train_full = X_train_full.reshape(X_train_full.shape[0], -1)
    X_test = X_test.reshape(X_test.shape[0], -1)

    # Normaliza莽茫o (escala de 0 a 1)
    X_train_full = X_train_full.astype('float32') / 255.0
    X_test = X_test.astype('float32') / 255.0

    # Separa莽茫o Treino e Valida莽茫o
    X_train, X_val, y_train, y_val = train_test_split(
        X_train_full, y_train_full, test_size=0.2, random_state=seed
    )

    return X_train, X_val, y_train.flatten(), y_val.flatten()

# Execu莽茫o
X_train, X_val, y_train, y_val = load_data(seed=42)
print("Dados carregados com sucesso!")
print(f"Shape de X_train: {X_train.shape}")
print(f"Shape de X_val: {X_val.shape}")
````

Formato original: $32 \times 32 \times 3$ (Altura, Largura, Canais RGB).


Features ap贸s flatten: $3072$ features ($32 \times 32 \times 3$).

Por que o flatten 茅 necess谩rio? Uma MLP (Multilayer Perceptron) possui uma camada de entrada composta por neur么nios unidimensionais. Ela n茫o consegue processar a estrutura espacial 2D/3D diretamente; cada pixel deve ser uma entrada independente.

Import芒ncia da normaliza莽茫o: Facilita a converg锚ncia do Gradiente Descendente. Sem ela, pixels com valores altos (255) causariam atualiza莽玫es de pesos muito bruscas, tornando o treinamento inst谩vel ou muito lento.

## Requisitos

Sua implementa莽茫o deve:

- utilizar `MLPClassifier` do `sklearn`;
- permitir diferentes arquiteturas atrav茅s do par芒metro `hidden_layers`;
- utilizar:
  - `activation`
  - `learning_rate`
  - `random_state`
- treinar o modelo utilizando `fit`.

A fun莽茫o deve retornar o modelo treinado.

Al茅m disso, responda:

1. Quantos par芒metros existem na primeira camada?
2. Qual a fun莽茫o da ativa莽茫o ReLU?
3. Por que MLPs possuem muitos par芒metros ao trabalhar com imagens?

**Solu莽茫o**:

````python
from sklearn.neural_network import MLPClassifier

def train_mlp(X_train, y_train, activation='relu', hidden_layers=(100,), learning_rate=0.001, seed=42):

    model = MLPClassifier(
        hidden_layer_sizes=hidden_layers,
        activation=activation,
        learning_rate_init=learning_rate,
        random_state=seed,
        max_iter=30
    )

    model.fit(X_train, y_train)

    return model

print("Treinando a MLP... Por favor, aguarde.")
meu_modelo = train_mlp(X_train, y_train, activation='relu', hidden_layers=(128,), learning_rate=0.001, seed=42)
print("Sucesso! Modelo treinado.")
print(f"Configura莽茫o do modelo: {meu_modelo}")
````

Quantos par芒metros existem na primeira camada?

A quantidade de par芒metros 茅 calculada pela f贸rmula: $(N_{features} \times N_{neuronios}) + N_{biases}$.No CIFAR-10, temos 3072 features de entrada. Se usarmos uma primeira camada com 128 neur么nios, teremos:$(3072 \times 128) + 128 = 393.344$ par芒metros apenas na primeira conex茫o.

Qual a fun莽茫o da ativa莽茫o ReLU?

A fun莽茫o ReLU ($f(x) = \max(0, x)$) atua como um filtro que decide se o neur么nio deve "disparar" ou n茫o. Ela zera valores negativos e mant茅m valores positivos. Sua principal fun莽茫o 茅 permitir o treinamento de redes profundas sem que o sinal do erro desapare莽a (evita o vanishing gradient), sendo muito mais r谩pida de computar que a Sigmoide ou Tanh.


Por que MLPs possuem muitos par芒metros ao trabalhar com imagens?

Porque a MLP 茅 uma rede "totalmente conectada" (Fully Connected). Isso significa que cada 煤nico pixel da imagem precisa ter um peso (conex茫o) para cada neur么nio da pr贸xima camada. Como imagens possuem muitos pixels (3072 no caso do CIFAR-10), o n煤mero de conex玫es cresce de forma explosiva, diferente de redes convolucionais que "reutilizam" pesos.

# Quest茫o 3

Implemente a fun莽茫o:

```python
evaluate(model, X_test, y_test)
```

Ela deve:

- realizar predi莽玫es;
- calcular:
  - accuracy;
  - precision;
  - recall;
  - f1-score.

Utilize `sklearn.metrics`.

Al茅m disso:

- apresente os resultados em um dicion谩rio ou DataFrame;
- interprete os resultados obtidos.

Responda:

1. O que a accuracy representa?
2. Qual a diferen莽a entre precision e recall?
3. Em quais situa莽玫es o f1-score 茅 importante?

**Solu莽茫o**:

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd

def evaluate(model, X_test, y_test):
    y_pred = model.predict(X_test)

    metrics = {
        "accuracy": accuracy_score(y_test, y_pred),
        "precision": precision_score(y_test, y_pred, average='macro'),
        "recall": recall_score(y_test, y_pred, average='macro'),
        "f1_score": f1_score(y_test, y_pred, average='macro')
    }

    df_results = pd.DataFrame([metrics])

    return df_results

print("Avaliando o modelo nos dados de valida莽茫o...")
resultados_df = evaluate(meu_modelo, X_val, y_val)

display(resultados_df)
````

O que a accuracy representa?

A acur谩cia 茅 a m茅trica mais simples: ela representa a porcentagem total de acertos do modelo. Se a acur谩cia 茅 0.40, significa que o modelo classificou corretamente 40% das imagens do dataset.

Qual a diferen莽a entre precision e recall?

Precision (Precis茫o): Foca na qualidade da predi莽茫o positiva. Responde: "De todas as imagens que o modelo disse ser 'Cachorro', quantas eram realmente 'Cachorro'?". 脡 importante para evitar falsos positivos.

Recall (Revoca莽茫o): Foca na quantidade de acertos de uma classe. Responde: "De todas as imagens de 'Cachorro' que existiam no dataset, quantas o modelo conseguiu encontrar?". 脡 importante para evitar falsos negativos.

Em quais situa莽玫es o f1-score 茅 importante?

O F1-Score 茅 a m茅dia harm么nica entre precis茫o e recall. Ele 茅 indispens谩vel quando temos um dataset desbalanceado (ex: muito mais fotos de carros do que de gatos) ou quando queremos um equil铆brio real entre precis茫o e recall, garantindo que o modelo n茫o esteja sacrificando um pelo outro.

Responda:

2. Qual configura莽茫o apresentou maior estabilidade?

**Solu莽茫o**:

````python
import mlflow
import time

mlflow.set_tracking_uri("file:///content/mlruns")
mlflow.set_experiment("CIFAR10_Tracking")

def run_experiment(run_name, activation, hidden_layers, lr):
    with mlflow.start_run(run_name=run_name):

        mlflow.log_param("activation", activation)
        mlflow.log_param("hidden_layers", str(hidden_layers))
        mlflow.log_param("learning_rate", lr)
        mlflow.log_param("max_iter", 30)
        mlflow.log_param("batch_size", "auto")

        start_time = time.time()
        model = train_mlp(X_train, y_train, activation, hidden_layers, lr)
        duration = time.time() - start_time

        res_df = evaluate(model, X_val, y_val)


        mlflow.log_metric("accuracy", res_df["accuracy"].values[0])
        mlflow.log_metric("precision", res_df["precision"].values[0])
        mlflow.log_metric("recall", res_df["recall"].values[0])
        mlflow.log_metric("f1_score", res_df["f1_score"].values[0])
        mlflow.log_metric("training_time", duration)

        print(f"Execu莽茫o '{run_name}' salva no MLflow!")

run_experiment("Teste_Inicial", "relu", (128, 64), 0.001)

print("\n--- Tabela de Experimentos (MLflow UI alternativa) ---")
display(mlflow.search_runs())
````

1. Qual experimento apresentou melhor desempenho?

    Teste inicial

2. Qual configura莽茫o apresentou maior estabilidade?

    Baseado nos par芒metros registrados, a configura莽茫o com Learning Rate de 0.001 e ativa莽茫o ReLU apresentou maior estabilidade. Isso 茅 evidenciado pelo fato de o modelo ter conseguido convergir para uma acur谩cia pr贸xima de 48% sem divergir

3. Qual o benef铆cio do rastreamento experimental?

    Compara莽茫o Objetiva: Voc锚 n茫o precisa "adivinhar" qual modelo foi melhor; os n煤meros est茫o lado a lado.

    eprodutibilidade: Qualquer pessoa pode repetir exatamente o mesmo teste usando os mesmos par芒metros registrados.
